```python
import tensorflow as tf
from keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output, HTML
```

```python
gan_generator = load_model("./model/cifar_generator_test.h5")
gan_generator.summary()

# class names for CIFAR-10
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
for i in range(10):
    # creating noise
    latent_z = np.random.normal(size=(100, 100))
    
    imgs = gan_generator.predict(latent_z)
    
    fig = plt.figure(figsize=(20, 20), tight_layout=True)
    for idx, img in enumerate(imgs):
        ax = fig.add_subplot(10, 10, idx+1)
        ax.imshow((img + 1) / 2)
        ax.axis('off')
```
